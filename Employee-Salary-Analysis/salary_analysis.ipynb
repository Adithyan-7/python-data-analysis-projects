```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("employee_salary.csv")
df_index=df.set_index("Employee_ID")
print(df_index)

grp = df.groupby("Department")["Salary"]

highest_employee = df.loc[df["Salary"].idxmax()]
lowest_employee = df.loc[df["Salary"].idxmin()]

print("\n\nHighest paid employee \n",highest_employee)
print("\n\nLowest paid employee \n",lowest_employee)

average_salary = grp.mean()
print("\nAverage salary:", average_salary)

std_salary=df["Salary"].std()
print("\nSalary Standard Deviation:",std_salary)

dept_size=df.groupby("Department").size()
print(dept_size)

plt.bar(average_salary.index, average_salary.values)
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Department-wise Average Salary")
plt.tight_layout()
plt.show()

plt.figure()
plt.boxplot(df["Salary"])
plt.title("Salary Distribution")
plt.ylabel("Salary")


plt.figure()
plt.pie(dept_size.values, labels=dept_size.index)
plt.title("Department Size Distribution")
plt.show()
```
