```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = { "Name": ["Arun", "Athul", "Abhi", "David", "Ahammed",
             "Farddheen", "Gokul", "Hema", "Ishan", "Malu"],
    
    "Maths": [85, 78, 92, 70, 88, 95, 60, 73, 81, 90],
    "Physics": [80, 75, 89, 65, 84, 91, 58, 72, 79, 87],
    "Chemistry": [82, 74, 90, 68, 86, 93, 55, 70, 77, 88],
    "English": [88, 80, 85, 72, 90, 94, 65, 75, 83, 89]}

df = pd.DataFrame(data)
print(df)

df["Total"] = df["Maths"] + df["Physics"] + df["Chemistry"] + df["English"]
df["Average"]=df["Total"]/4

highest=df["Average"].max()
lowest=df["Average"].min()

print(df[["Name", "Total", "Average"]])
print("Highest of Average",highest)
print("Lowest of Average",lowest)

avg_subject=df[[ "Maths", "Physics", "Chemistry","English"]].mean()
print("Average of all the subject:\n",avg_subject)

plt.figure()
plt.bar(avg_subject.index, avg_subject.values)
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Subject-wise Average Marks")
plt.show()

plt.figure()
corr = df[["Maths", "Physics", "Chemistry", "English"]].corr()
sns.heatmap(corr,cmap="coolwarm",annot=True)
plt.title("Subject Correlation Heatmap")
plt.show()

plt.figure()
sns.boxplot(data=df[["Maths", "Physics", "Chemistry", "English"]])
plt.title("Marks Distribution")
plt.show()

top= df.nlargest(3, "Total")
print(top[["Name", "Total", "Average"]])
```
