```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('UberDataset.csv')
df
```

```python
df.info()
```

```python
df['PURPOSE'].fillna("Not Available",inplace=True)
df
```

```python
df['STARTDATE']=pd.to_datetime(df['STARTDATE'],errors='coerce')
df
df['ENDDATE']=pd.to_datetime(df['ENDDATE'],errors='coerce')
df
```

```python
df['date']=pd.to_datetime(df['STARTDATE']).dt.date
df['time']=pd.to_datetime(df['STARTDATE']).dt.hour

df['day-night']=pd.cut(x=df['time'],bins=[0,10,15,19,24],labels=['Morning','Afternoon','Evening','Night'],right=False)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.countplot(x=df['CATEGORY'])
plt.xticks(rotation=90)

plt.subplot(1,2,2)
sns.countplot(x=df['PURPOSE'])
plt.xticks(rotation=90)
```

```python
sns.countplot(x=df['day-night'])
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='PURPOSE',hue='CATEGORY')
plt.xticks(rotation=90)
plt.show()
```

```python
numeric_df=df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(),cmap='BrBG',fmt='.2f',linewidth=2,annot=True)
```
